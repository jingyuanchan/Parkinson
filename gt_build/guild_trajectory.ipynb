```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from glob import glob

csv_gt = r'C:\Users\jingy\Parkinson\Codes\csv\FW_gt.csv'
df_FW = pd.read_csv(csv_gt)
# df_FW['trajectory']  = np.nan
trajectory_list=[]
traj_len_list = []
name_list = []
for index, row in df_FW.iterrows():
    npy_dir = r'C:\Users\jingy\Parkinson\Codes\data\keypoints_03_19\FW/'
    video_name= row['video']
    file_name = npy_dir+ video_name[:-3]+'npy'
    target_id = row['id']
    name_id = video_name.split('_')[0]
    # make sure npy_file for each video exists
    try:
        os.path.exists(file_name) == True
    except:
        print(file_name)

    # Creat any empty list to store the traj for each id 
    traj_dict = {}
    read_dict = np.load(file_name, allow_pickle=True).item()
    # itterate by frame index
    for frame_index in read_dict:
        # itterate id within a frame result
        for id in read_dict[frame_index]:
            # append the id as a key to the traj_dict
            if id not in traj_dict.keys():
                traj_dict[id]=[]

            pts = read_dict[frame_index][id]
            # append the obtain the pts within the frame to form a traj
            traj_dict[id].append(pts)
    
    traget_traj = traj_dict[str(target_id)]
    trajectory_list.append(traget_traj)
    traj_len_list.append(len(traget_traj))
    name_list.append(name_id)
    
df_FW['name_code'] = name_list
df_FW['trajectory'] = trajectory_list
df_FW['traj_length'] = traj_len_list
```
